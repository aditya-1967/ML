```python
import pandas as pd
import numpy as np
import seaborn as sns
```

### Near Earth Objects

```python
raw_df = pd.read_csv('neo.csv')
```

```python
raw_df.head()
```

```python
raw_df.drop(['id', 'name', 'orbiting_body', 'sentry_object'], inplace=True, axis=1)
```

```python
raw_df.head()
```

```python
raw_df['hazardous'] = raw_df['hazardous'].map({False:0, True:1})
```

```python
raw_df.head()
```

```python
raw_df.corr()
```

```python
raw_df.isnull().sum()
```

```python
raw_df.drop(['absolute_magnitude'], axis=1, inplace= True)
```

```python
raw_df.head()
```

```python
raw_df['miss_distance']
```

```python
raw_df.corr()
```

```python
sns.heatmap(raw_df.corr())
```

```python
X = raw_df.iloc[:,:-1]
Y = raw_df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state= 42)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'penalty':['l1', 'l2', 'elasticnet'], 'C':[1,2,3,4,5,8,10,20,30,40,50], 'max_iter':[100,200,300],}
```

```python
model_new = GridSearchCV(model, param_grid = parameters, scoring = 'accuracy', cv = 5)
```

```python
model_new.fit(x_train, y_train)
```

```python
model_new.best_params_
```

```python
model_new.best_score_
```

```python
y_pred = model_new.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
```

```python
score *= 100
score
```

```python
report = classification_report(y_test, y_pred)
```

```python
report
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
```
