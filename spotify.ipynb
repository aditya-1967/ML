```python
raw_df = pd.read_csv('/kaggle/input/top-hits-spotify-from-20002019/songs_normalize.csv')
raw_df.head()
```

```python
raw_df.isna().sum()
```

```python
raw_df.info()
```

```python
raw_df['explicit'] = raw_df['explicit'].map({False: 0, True: 1})
```

```python
raw_df.head()
```

```python
raw_df.drop(['artist', 'song'], axis = 1, inplace = True)
```

```python
raw_df.corr()
```

```python
raw_df.genre.unique()
```

```python
genre = pd.get_dummies(raw_df['genre'])
raw_df_new = pd.concat([raw_df, genre], axis = 1)
```

```python
raw_df_new.head()
```

```python
raw_df_new.info()
```

```python
raw_df_new.drop('genre', axis = 1, inplace = True)
```

```python
x = raw_df_new.drop('popularity', axis = 1)
y = raw_df_new['popularity']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
